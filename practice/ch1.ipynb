### 1.6 matplotlib

#### 1.6.1. 단순 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격으로 생성
y = np.sin(x)

# 그래프
plt.plot(x,y)
plt.show()
```

#### pyplot의 기능

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)


# 그래프
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle = "--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()
plt.show()
```
